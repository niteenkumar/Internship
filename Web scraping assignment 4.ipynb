```python
#Make empty lists
title = []   
series = []
place = []
date = []
times = []

    
# Extracting series
try:
    ser = driver.find_elements_by_xpath("//div[@class='fixture-card-top']/h5/span")      
    for i in ser:
        series.append(i.text)
except NoSuchElementException:
    series.append('-')    

series=series[1::2]

# Extracting match title and place
try:
    places = driver.find_elements_by_xpath("//div[@class='fix-place ng-binding ng-scope']")      
    for i in places:
        title.append(i.text.split('-')[0])
        place.append(i.text.split('-')[1])
except NoSuchElementException:
    place.append('-')


time.sleep(1)    
# Extracting date    
try:
    dates = driver.find_elements_by_xpath("//div[@class='match-card-left match-schedule']")      
    for i in dates:
        date.append(i.text)
except NoSuchElementException:
    date.append('-')


time.sleep(1)
# Extracting time
try:
    t = driver.find_elements_by_xpath("//div[@class='match-card-right match-schedule ']/h5")      
    for i in t:
        times.append(i.text)
except NoSuchElementException:
    times.append('-')



time.sleep(2)    
# Make dataframe
fixture = pd.DataFrame({'Match Title':title,'Series':series,'Place':place,'Date':date,'Time':times})
fixture.head()
```

```python
# reaching to ShowAllDataset page
title_tags=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td/span/b/a")
driver.get(title_tags.get_attribute("href"))
```

```python
#Make empty lists
name = []   
dat = []
task = []
typ = []
instance = []
attr = []
year = []

urls = []
url=driver.find_elements_by_xpath("//p[@class='normal']/b/a")
for i in url:
    urls.append(i.get_attribute("href"))
    
for i in urls:
    driver.get(i)
    time.sleep(2)
    
    # Extracting Dataset name
    try:
        des = driver.find_element_by_xpath("//span[@class='heading']/b")      
        name.append(des.text)
    except NoSuchElementException:
        name.append('-') 
        
    # Extracting Data type
    try:
        c = driver.find_element_by_xpath("//table[@border='1']/tbody/tr/td[2]")      
        dat.append(c.text)
    except NoSuchElementException:
        dat.append('-') 
        
    # Extracting Task
    try:
        l = driver.find_element_by_xpath("//table[@border='1']/tbody/tr[3]/td[2]")
        task.append(l.text)
    except NoSuchElementException:
        task.append('-') 
        
    # Extracting Attribute type
    try:
        l = driver.find_element_by_xpath("//table[@border='1']/tbody/tr[2]/td[2]")
        typ.append(l.text)
    except NoSuchElementException:
        typ.append('-')
        
    # Extracting No of instances
    try:
        l = driver.find_element_by_xpath("//table[@border='1']/tbody/tr/td[4]")
        instance.append(l.text)
    except NoSuchElementException:
        instance.append('-')
        
    # Extracting No of attribute
    try:
        l = driver.find_element_by_xpath("//table[@border='1']/tbody/tr[2]/td[4]")
        attr.append(l.text)
    except NoSuchElementException:
        attr.append('-')
        
    # Extracting Year
    try:
        l = driver.find_element_by_xpath("//table[@border='1']/tbody/tr[2]/td[6]")
        year.append(l.text)
    except NoSuchElementException:
        year.append('-')

time.sleep(2)    
# Make dataframe
Data = pd.DataFrame({'Dataset name':name,'Data type':dat,'Task':task,'Attribute type':typ,'No of instances':instance,'No of attribute':attr,'Year':year})
Data.head()
```
